```python
import pandas as pd
!pip install odfpy
dataset_composition_file = 'dataset_composition.ods'
```

```python
# read the 3 tables of the excel file
train_dataset = pd.read_excel(dataset_composition_file,
sheet_name='Train',
header=0,
index_col=0,
keep_default_na=True
)

test_dataset = pd.read_excel(dataset_composition_file,
sheet_name='Test',
header=0,
index_col=0,
keep_default_na=True
)

patients_info = pd.read_excel(dataset_composition_file,
sheet_name='Patients',
header=0,
index_col=0,
keep_default_na=True
)
```

###Patients

```python
# Patients distribution, barplot
patients_distribution = patients_info.plot.bar(title='Patients distribution in the dataset')
for p in patients_distribution.patches:
    patients_distribution.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
# Patients distribution, pie chart
patients_distribution_pie_train = patients_info.plot.pie(subplots=True, figsize=(12,6), title='Patients in the train/test datasets', fontsize=11)
```

### Images

```python
images_sum = pd.concat([train_dataset.sum(axis=1),test_dataset.sum(axis=1)], axis=1)
images_sum.columns=['Train', 'Test']
images_sum_plot = images_sum.plot.bar(xlabel='Nucleus', title='Number of segmented cells in the training set')
for p in images_sum_plot.patches:
    images_sum_plot.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

train_dataset.plot.bar(xlabel='Nucleus', ylabel='Frequency', title='Distribution of the training samples')
test_dataset.plot.bar(xlabel='Nucleus', ylabel='Frequency', title='Distribution of the testing samples')
```
